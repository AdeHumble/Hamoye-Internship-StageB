## **3. Check data type**
- I would like to check if every column in the dataset contains a unique(the same) data types
- In order not to waste time by using the dtype() on each column, I will prefer to use a for-loop statement to perform this task across the whole dataset columns at once

```python
# Create an empty DataFrame to store the results
data_type_df = pd.DataFrame(columns=['COLUMN NAMES', 'UNIQUE DATA TYPES'])

# Iterate through columns and store unique data types in the new DataFrame
for column_name in data.columns:
    unique_data_types = data[column_name].apply(type).unique()
    
    #data_type_df = data_type_df.append({'COLUMN NAME': column_name, 'UNIQUE DATA TYPES': unique_data_types}, ignore_index=True)
    data_type_df = pd.concat([data_type_df, pd.DataFrame({'COLUMN NAMES': [column_name], 'UNIQUE DATA TYPES': [unique_data_types]})],ignore_index=True)

# Display the resulting table
print(data_type_df)
```

### **INFERENCE:**
- Each column contains a unique data type, EXCEPT for date column. Hence, there would be no need to perform type casting

```python
data['date'] = pd.to_datetime(data['date'])
print(data_type_df)
```

### **INFERENCE:**
- Now, all the columns in the dataset have the proper data types

## **1. Splitting the data into train and test data**

```python
# 'X' contain the feature columns, and 'y' is the target variable (Appliances energy usage).
X = data.drop(columns=['date','lights','Appliances'])  # Feature columns
y = data['Appliances']  # Target variable

# Split the data into a training set and a testing set (using 70% training and 30% testing).
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Hence, for information on the structure of the splitted data
print(f"The training data has {len(X_train)} rows")
print(f"The testing data has {len(X_test)} rows")
```
